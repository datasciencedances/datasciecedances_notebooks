```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def prob(N):
    ans = []
    for r in range(1, N+1):
        ans.append(r/N*np.log(N/r))
    return ans, list(range(1, N+1))
```

```python
def prob2(N):
    ans = []
    for r in range(1, N+1):
        a = 0
        for i in range(r, N):
            a+=1/i
        ans.append(a*r/N)
    return ans, list(range(1, N+1))
```

```python
plt.rcParams.update({'font.size': 20})

fig, axes = plt.subplots(1,2, figsize=(20,7))
p, r = prob(10)
print(p)
sns.scatterplot(x=r, y=p,ax=axes[0],s=100)
axes[0].set_xticks(r,r)
axes[0].text(x=r[3],y=p[3], s=round(p[3],3))
axes[0].set_ylim(0, 0.45)
axes[0].grid()
axes[0].set_xlabel('r')
axes[0].set_title('Sử dụng công thức tích phân')
p, r = prob2(10)
sns.scatterplot(x=r, y=p, ax=axes[1], s=100)
axes[1].set_xticks(r,r)
axes[1].text(x=r[2],y=p[2], s=round(p[2],3))
axes[1].set_ylim(0, 0.45)
axes[1].grid()
axes[1].set_xlabel('r')
axes[1].set_title('Sử dụng công thức cộng')
plt.suptitle("Số lượng người (r) cần nói chuyện(bỏ qua) để tìm được gu \n và xác suất để tìm được người tốt nhất")
plt.tight_layout()
```
